```python
# ══════════════════════════════════════════════════════════
# MUSTERLÖSUNG — Aufgabe 1a: Statistiken
# ══════════════════════════════════════════════════════════

stats = {}
for name, data in anscombe.items():
    x = np.array(data['x'])
    y = np.array(data['y'])
    r = np.corrcoef(x, y)[0, 1]
    stats[name] = {
        'mean_x':   round(np.mean(x), 2),
        'mean_y':   round(np.mean(y), 2),
        'var_x':    round(np.var(x, ddof=1), 2),
        'var_y':    round(np.var(y, ddof=1), 2),
        'corr_r':   round(r, 3),
    }

df_stats = pd.DataFrame(stats).T
print('Anscombe Quartet — Statistiken:')
print(df_stats.to_string())
```

```python
# ══════════════════════════════════════════════════════════
# MUSTERLÖSUNG — Aufgabe 1b: 2x2 Scatterplot
# ══════════════════════════════════════════════════════════

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

for idx, (name, data) in enumerate(anscombe.items()):
    x = np.array(data['x'])
    y = np.array(data['y'])
    m, b = np.polyfit(x, y, 1)
    x_line = np.linspace(x.min() - 1, x.max() + 1, 100)

    ax = axes[idx]
    ax.scatter(x, y, color=TEAL, s=80, zorder=5, edgecolors='white', linewidth=1)
    ax.plot(x_line, m * x_line + b, color=SLATE, linewidth=1.5, linestyle='--', alpha=0.7)
    ax.set_title(f'Dataset {name}', fontsize=13, fontweight='bold', color=NAVY)
    ax.set_xlabel('x', fontsize=11)
    ax.set_ylabel('y', fontsize=11)
    ax.text(0.05, 0.92, f'r = {np.corrcoef(x,y)[0,1]:.3f}',
            transform=ax.transAxes, fontsize=9.5, color=SLATE)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.yaxis.grid(True, linestyle=':', linewidth=0.7, color='#CBD5E1', zorder=0)

fig.suptitle("Anscombe's Quartet — gleiche Statistiken, völlig verschiedene Muster",
             fontsize=14, fontweight='bold', color=NAVY, y=1.01)
plt.tight_layout()
plt.show()
```

```python
# ══════════════════════════════════════════════════════════
# MUSTERLÖSUNG — Aufgabe 2a: Vergleich Umsatz nach Kanal
# ══════════════════════════════════════════════════════════

# Sortieren — für Vergleiche ist absteigende Sortierung intuitiver
sortiert = sorted(zip(umsatz, kanaele), reverse=True)
ums_sort, kan_sort = zip(*sortiert)

# Highlight-Strategie: höchster Balken in Akzentfarbe
farben = [TEAL if u == max(ums_sort) else '#CBD5E1' for u in ums_sort]

fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.barh(kan_sort, ums_sort, color=farben, height=0.6)

# Direkte Datenbeschriftung
for bar, val in zip(bars, ums_sort):
    ax.text(val + 0.05, bar.get_y() + bar.get_height() / 2,
            f'{val:.1f} Mio.', va='center', fontsize=11, color=NAVY)

ax.set_title('Stationärer Handel führt, aber Online holt auf',
             fontsize=13, fontweight='bold', color=NAVY, pad=10)
ax.set_xlabel('Umsatz in Mio. CHF', fontsize=11)
ax.set_xlim(0, 8.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.grid(False)
ax.xaxis.grid(True, linestyle=':', linewidth=0.7, color='#E2E8F0', zorder=0)
ax.set_axisbelow(True)
plt.tight_layout()
plt.show()
```
